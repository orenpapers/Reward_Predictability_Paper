```python
import seaborn as sns
from shared_analysis_utils import strip_est, strip_fs
import matplotlib.pyplot as plt

msc_df["est"] = strip_est(msc_df["est"])
msc_df["fs"] = strip_fs(msc_df["fs"])
```

```python


msc_df_scrambled = msc_df[msc_df.cond=='scrambled']
msc_df_intact = msc_df[msc_df.cond=='intact']
df_to_viz_it = msc_df_intact[(msc_df_intact.fs.isin(["PCA(50)"])) & (msc_df_intact.est.isin(["SVC('poly')"]))].reset_index(drop=True)

df_to_viz_sc = msc_df_scrambled[(msc_df_scrambled.fs.isin(["PCA(50)"])) & (msc_df_scrambled.est.isin(["SVC('poly')"]))].reset_index(drop=True)
df_to_viz_sc
```

```python
from shared_analysis_utils import draw_stars_box
import random
random.seed(42)
l = len(df_to_viz_sc[df_to_viz_sc.train_type=='actual'])

fig, axs = plt.subplots(1, 2, figsize=(12,4))
[ax.set_ylim([0,0.1]) for ax in axs]


print(f"Scrambled Acc is {df_to_viz_sc[df_to_viz_sc.train_type=='actual'].accuracy.mean()}")
print(f"Intact Acc is {df_to_viz_it[df_to_viz_it.train_type=='actual'].accuracy.mean()}")
# print(f"Vision Acc is {df_to_viz_vs[df_to_viz_vs.train_type=='actual'].accuracy.mean()}")


draw_stars_box(df_to_viz_sc[df_to_viz_sc.train_type=='actual'].accuracy,
               random.sample(list(df_to_viz_sc[df_to_viz_sc.train_type=='shuffled'].accuracy), l),
               "wilcoxon", title= "Scrambled - Reward", suf = "Scrambled" , ax=axs[1], show=False,
               ylabel="Accuracy")


draw_stars_box(df_to_viz_it[df_to_viz_it.train_type=='actual'].accuracy,
               random.sample(list(df_to_viz_it[df_to_viz_it.train_type=='shuffled'].accuracy), l),
               "wilcoxon", title= "Intact - Reward", suf = "Intact", ax=axs[0], show=False,
               ylabel="Accuracy")


```
