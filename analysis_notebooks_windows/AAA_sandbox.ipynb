```python
import joblib
from configs.params import alignment_dict_fn_onlyaim2 as alignment_dict_fn
alignment_dict = joblib.load(alignment_dict_fn)
nlp_df = alignment_dict["nlp_data_df"]
nlp_df
```

```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
sid_obj = SentimentIntensityAnalyzer()
l1 = [sid_obj.polarity_scores(x)['neu'] for x in nlp_df['baseline_text']]
l2 = [sid_obj.polarity_scores(x)['pos'] for x in nlp_df['baseline_text']]
l3 = [sid_obj.polarity_scores(x)['neg'] for x in nlp_df['baseline_text']]
```

```python
from matplotlib import pyplot as plt
plt.figure(figsize=(9,6))
plt.plot(l1, color='r', label='Neutrality score')
plt.plot(l2, color='b', label='Positivity score')
plt.plot(l3, color='g', label='Negetivity score')
plt.xlabel("sentence idx")
plt.ylabel("Score")
plt.legend()                                                                                                           
```
