```python
from shared_analysis_utils import draw_stars_box

for x3 in ['reward','vision']:

    # plt.figure(figsize=(12,4))
    fig, axs = plt.subplots(1, 2, figsize=(12,4))
    # fig.suptitle(f"{x3}", fontsize=16)

    df_to_viz2 = df_to_viz[df_to_viz.roi == x3]

    sns.histplot(df_to_viz2[df_to_viz2.cond == 'scrambled']['mse'],  bins=20, ax = axs[0])
    actual_dist = df_to_viz2[df_to_viz2.cond == 'intact']['mse']
    actual_corr = actual_dist.mean()
    axs[0].axvline(x=actual_corr, color='green')
    axs[0].set_xlabel('MSE', fontsize=14)
    axs[0].set_ylabel('Count', fontsize=14)
    null_dist = df_to_viz2[df_to_viz2.cond == 'scrambled']['mse']

    actual_rank = [x for x in null_dist if x < actual_corr]

    # plt.show()
    axs[1].set_title("Per fold MSE")
    axs[1].set_ylabel("MSE", fontsize=14)
    s,p = draw_stars_box(actual_dist, null_dist, "wilcoxon", f"{x3}", ax=axs[1], alternative='less', verbose=False, ylabel="MSE",
                         xlabels = ['intact','scrambled'], title = f"{x3} - per fold MSE")

    print(f"For {x3}: Total {len(null_dist)} random vec, rank of actual ({actual_corr:.2f}) is {len(actual_rank)/len(null_dist):.2f} ({len(actual_rank)}/{len(null_dist)})")
    print(s)
    plt.show()
```
