```python
df_to_viz = explode_df[(explode_df.fs == "PCA(10)") & (explode_df.est == "SVC('linear')")]
df_to_viz
```

```python
df_to_viz['roi'] = [x.capitalize() for x in df_to_viz['roi']]
df_to_viz['model_type'] = [x.capitalize() for x in df_to_viz['model_type']]
sns.boxplot(data=df_to_viz.rename({"acc_vec": "Accuracy", "cond" : "Condition"}, axis=1),  x='roi', hue='Condition',   y='Accuracy')
```

```python
from scipy.stats import wilcoxon
for roi_str, roi_df in df_to_viz.groupby("roi"):
    print(roi_str)
    null_dist = roi_df[roi_df.cond=='scrambled'].acc_vec
    actual_dist = roi_df[roi_df.cond=='intact'].acc_vec
    actual_error = actual_dist.mean()
    actual_rank = [x for x in null_dist if x > actual_error]

    print(f"{roi_str}: Total {len(null_dist)} random vec, rank of actual ({actual_error}) is {len(actual_rank)/len(null_dist)} ({len(actual_rank)}/{len(null_dist)})")

    print("Wilcoxon ", wilcoxon(actual_dist, null_dist, alternative='greater'))
```
