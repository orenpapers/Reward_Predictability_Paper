```python
import joblib
from configs.params import alignment_dict_fn_onlyaim2 as alignment_dict_fn
alignment_dict = joblib.load(alignment_dict_fn)
nlp_df = alignment_dict["nlp_data_df"]
nlp_df
```

```python
df_to_r_check = pd.merge(left=df_to_r_check_exploded, right=nlp_df[['sentence_idx','vodka_perplexity_score']], left_on='test_sentence',right_on='sentence_idx')
df_to_r_check
```

```python
from scipy.stats import pearsonr, spearmanr, kendalltau
import scipy.stats as ss
d = {}

print("=== On Scaled: ===")

from sklearn.preprocessing import RobustScaler, StandardScaler
for conf,conf_df in df_to_r_check.groupby(["cond","train_type","fs","est"]):

    per_sentence_accuracy =  StandardScaler().fit_transform(conf_df.groupby('test_sentence').success.mean().values.reshape(-1,1))[:,0].tolist()
    per_sentence_perplexity = StandardScaler().fit_transform(conf_df.groupby('test_sentence').vodka_perplexity_score.mean().values.reshape(-1,1))[:,0].tolist()
    r1,p1 = pearsonr(per_sentence_accuracy, per_sentence_perplexity)
    r2,p2 = spearmanr(per_sentence_accuracy, per_sentence_perplexity)
    r3,p3 = kendalltau(per_sentence_accuracy, per_sentence_perplexity)
    if p1 < 0.1 or r3 > 0.2:
        print(f"Pearson: for {conf} got (r={r1:.2f}, p={p1:.7f})")# =  ({len(per_sentence_accuracy)}, {len(per_sentence_accuracy)})")
    if p2 < 0.1 or r2 > 0.2:
        print(f"Spearman: for {conf} got (r={r2:.2f}, p={p2:.7f})")# =  ({len(per_sentence_accuracy)}, {len(per_sentence_accuracy)})")
    if p3 < 0.1 or r3 > 0.2:
        print(f"Kandell: for {conf} got (r={r3:.2f}, p={p3:.7f})")# =  ({len(per_sentence_accuracy)}, {len(per_sentence_accuracy)})")

print("=== On ordinal: ===")
for conf,conf_df in df_to_r_check.groupby(["cond","train_type","fs","est"]):

    per_sentence_accuracy = ss.rankdata(conf_df.groupby('test_sentence').success.mean()).tolist()
    per_sentence_perplexity = ss.rankdata(conf_df.groupby('test_sentence').vodka_perplexity_score.mean()).tolist()
    r1,p1 = pearsonr(per_sentence_accuracy, per_sentence_perplexity)
    r2,p2 = spearmanr(per_sentence_accuracy, per_sentence_perplexity)
    r3,p3 = kendalltau(per_sentence_accuracy, per_sentence_perplexity)
    if p1 < 0.1 or r3 > 0.2:
        print(f"Pearson: for {conf} got (r={r1:.2f}, p={p1:.7f})")# =  ({len(per_sentence_accuracy)}, {len(per_sentence_accuracy)})")

    if p2 < 0.1 or r2 > 0.2:
        print(f"Spearman: for {conf} got (r={r2:.2f}, p={p2:.7f})")# =  ({len(per_sentence_accuracy)}, {len(per_sentence_accuracy)})")
    if p3 < 0.1 or r3 > 0.2:
        print(f"Kandell: for {conf} got (r={r3:.2f}, p={p3:.7f})")# =  ({len(per_sentence_accuracy)}, {len(per_sentence_accuracy)})")


print("=== On continouous: ===")
for conf,conf_df in df_to_r_check.groupby(["cond","train_type","fs","est"]):

    per_sentence_accuracy = conf_df.groupby('test_sentence').success.mean().tolist()
    per_sentence_perplexity = conf_df.groupby('test_sentence').vodka_perplexity_score.mean().tolist()
    r1,p1 = pearsonr(per_sentence_accuracy, per_sentence_perplexity)
    r2,p2 = spearmanr(per_sentence_accuracy, per_sentence_perplexity)
    r3,p3 = kendalltau(per_sentence_accuracy, per_sentence_perplexity)
    if p1 < 0.1 or r3 > 0.2 :
        print(f"Pearson: for {conf} got (r={r1:.2f}, p={p1:.7f})")# =  ({len(per_sentence_accuracy)}, {len(per_sentence_accuracy)})")
    if p2 < 0.1 or r2 > 0.2:
        tacc = per_sentence_accuracy
        tper = per_sentence_perplexity
        print(f"Spearman: for {conf} got (r={r2:.2f}, p={p2:.7f})")# =  ({len(per_sentence_accuracy)}, {len(per_sentence_accuracy)})")
    if p3 < 0.1 or r3 > 0.2:
        print(f"Kandell: for {conf} got (r={r3:.2f}, p={p3:.7f})")# =  ({len(per_sentence_accuracy)}, {len(per_sentence_accuracy)})")

```
